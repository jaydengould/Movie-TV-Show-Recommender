### Overview of dataset

**API:** https://developer.themoviedb.org/reference/intro/getting-started

Our dataset will be obtained from The Movie Database (TMDB) using their public API, which is free to access with a registered account. We will pull data from two endpoints: the movie endpoint and the TV show endpoint, collecting metadata on several thousand titles from each. The exact number of observations will depend on how many pages we pull from the API, but we are targeting around 5,000-10,000 movies and a similar number of TV shows.

Each record will include features like genres, cast, keywords, overview text, release date, runtime, original language, popularity score, and aggregate vote average and vote count. The variable types are a mix of categorical (genres, language, keywords), numerical (runtime, popularity, vote counts), and text (overview). Since we are building a content-based recommender, the most important features will be genres, keywords, and overview text.

Missing data is most common in less popular titles, particularly in the keywords and overview fields. We plan to drop records missing critical features or fill them where reasonable. We will pull and store the data locally as a CSV after the initial API call to avoid repeated requests during development.

### Overview of research question(s)

The research question we want answered is: given some user input related to movies/TV shows, are we able to identify similar movie titles based only on the content features?

The outcome variable is going to be a similarity score between titles, based on their content features, instead of basing it on a ratings system. There is no set outcome variable here because we are not predicting a specific variable. Our goal for this model is to rank other movie titles by similarity based on the user input and return some set number of recommendations.

The predictors we think are going to be the most useful are genres, keywords, and overview text. Genres and keywords are categorical features that capture what a title is about in general. Overview text adds more description and will likely be processed using some language processing method. Other features like original language, popularity, and runtime may help filter or re-rank recommendations, but are less crucial.

The goal of the model is to be descriptive. We are not trying to predict user behavior. We are describing the similarity structure of a large catalog of movies and TV shows and using that structure to create relevant recommendations. Another question we are interested in is whether recommendations tend to stay within the same genre and whether the patterns differ between movies and TV shows.

### Questions and Concerns

- Since we are building a recommender that covers both movies and TV shows, we will need to figure out if we should constrain recommendations to one type or be allowed to crossover.

- Overview text might not capture semantic similarity as well, where a movie or TV show can have different descriptions but can feel very similar while viewing.

- Another possible limitation is evaluation. Since our recommender does not predict a labeled response but instead ranks titles by similarity, there is no measure of accuracy. We therefore plan to evaluate the model through comparison with TMDB recommendations, overlap in major content features, and qualitative review of selected outputs.
